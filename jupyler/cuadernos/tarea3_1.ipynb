## Inciso a.

**Simular n = 1000 observaciones de una distribución exponencial con lambda = 2, usando 123 como semilla**

```python
import numpy
numpy.random.seed(123)
explambda = 2
datos = numpy.random.exponential(1/explambda, 1000)
datos
```

## Inciso b.
**Definir la función de verosimilitud negativa para lambda**

```python
def Ln(x):
   return numpy.log(x)

def verosimilitud_exponencial_negativa_logaritmica(explambda, datos):
   n = len(datos)
   return -(n*Ln(explambda) - explambda*numpy.sum(datos))
```

## Inciso c.
**Con la función minimize de scipy.optimize, sacar el estimador de máxima verosimilitud**

```python
import scipy
lambda_inicial=numpy.array([0.5])
resultao = scipy.optimize.minimize(verosimilitud_exponencial_negativa_logaritmica, x0=lambda_inicial, args=(datos,), method="Nelder-Mead")
lambda_gorrito = resultao.x[0]
print(f"lambda gorrito = {lambda_gorrito}")
print(f"Comparación con el promedio recíproco: {1/numpy.mean(datos)}")
```
