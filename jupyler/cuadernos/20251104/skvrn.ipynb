```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
wages = pd.read_csv('https://raw.githubusercontent.com/jealcalat/AEM-ITESO/refs/heads/main/datasets/wages.csv')
wages.head()
sns.lmplot(x="ed", y="earn", hue="sex", data=wages)
```

```python
wages["log_earn"] = np.log(wages["earn"])
sns.lmplot(x="ed", y="log_earn", hue="sex", data=wages)
```
